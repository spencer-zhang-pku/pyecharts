```python
import numpy as np
import pandas as pd
import json
from pandas.io.json import json_normalize
import pyecharts as pe
from collections import Counter
```

```python
op1=open(r'D:\try\shoes.json', 'r',encoding='utf-8')
li=[]
dict1={}
for i in op1:
    k=json.loads(i.encode("utf-8"))#把字符串转换为json
    li.append(k)
a=json_normalize(li)#把由json数据构成的列表转换成数据框
```

```python
a.columns
```

```python
a.sales=a.sales.str.split("人",expand=True)[0]
a.sales = a.sales.astype(np.int64)#转换列的类型为整数
a.price = a.price.astype(np.float)
```

```python
from pyecharts.globals import ThemeType
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.charts import Line
```

```python
#最简单的价格分布求取方式
#这种分布虽然现实的信息最详细，但是并不便于解读
p1=Counter(a.price.values)
x=list(p1.keys())
x.sort()
y=[p1[i] for i in x]
line1=Line()
line1.add("意尔康商品价格",x,y,xaxis_max=600)
```

value_count() 的参数：


(1) normalize : boolean, default False             如果为True，则返回的对象将包含唯一值的相对频率。

sort : boolean, default True             按值排序

ascending : boolean, default False        按频率计数升序排序

bins : integer, optional    而不是数值计算，把它们分成半开放的箱子，一个方便的pd.cut，只适用于数字数据

dropna : boolean, default True          不包括NaN的数量。

```python
import datetime
import random

from pyecharts import options as opts
from pyecharts.charts import Calendar


def calendar_base() -> Calendar:
    begin = datetime.date(2017, 1, 1)
    end = datetime.date(2017, 12, 31)
    data = [
        [str(begin + datetime.timedelta(days=i)), random.randint(1000, 25000)]
        for i in range((end - begin).days + 1)
    ]
    print(data[0:2])
    c = (
        Calendar()
        .add("", data, calendar_opts=opts.CalendarOpts(range_="2017"))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Calendar-2017年微信步数情况"),
            visualmap_opts=opts.VisualMapOpts(
                max_=20000,
                min_=500,
                orient="horizontal",
                is_piecewise=True,
                pos_top="230px",
                pos_left="100px",
            ),
        )
    )
    return c
```

```python
calendar_base().render_notebook()
```
